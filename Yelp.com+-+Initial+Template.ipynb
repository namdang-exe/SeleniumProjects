### Import Packages

```python
import selenium
from selenium import webdriver
import pandas as pd
import time
import numpy as np
```

### Set Up

```python
PATH = r"C:\Program Files (x86)\chromedriver.exe"
URL = "https://www.yelp.com/search?find_desc=Hot%20Fried%20Chicken&find_loc=Nashville%2C%20TN"
URL = "https://www.yelp.com/search?find_desc=Dentists&find_loc=San%20Francisco%2C%20CA"
driver = webdriver.Chrome(PATH)
driver.get(URL)
driver.maximize_window()
time.sleep(2)
```

### Accept Cookies

```python
try:
    cookies = driver.find_element_by_id('onetrust-accept-btn-handler')
    cookies.click()
except Exception:
    print("Moving on")
```

### Target The Data We Want To Scrape

```python
# Locators
# names
names = driver.find_elements_by_xpath('//h4[@class="css-1l5lt1i"]/span/a')
# address
addresses = driver.find_elements_by_xpath('//div[@class=" border-color--default__09f24__3Epto"]/p[@class="css-1j7sdmt"]/span[@class=" css-e81eai"]')
# expertise
expertises = driver.find_elements_by_xpath('//span[@class=" css-epvm6 display--inline__09f24__nqZ_W border-color--default__09f24__3Epto"]')
# stars
stars = driver.find_elements_by_xpath('//div[@class=" border-color--default__09f24__3Epto"]/div[@class=" attribute__09f24__1La1D display--inline-block__09f24__3SvIn margin-r1__09f24__3PebR border-color--default__09f24__3Epto"]/span/div')
# rating count
ratings = driver.find_elements_by_xpath('//div[@class=" attribute__09f24__1La1D display--inline-block__09f24__3SvIn border-color--default__09f24__3Epto"]/span')
```

### Empty Lists

1. Loop through the results
2. Append the results in a list
3. Use this lists as a value part for the dictionary (Remember: Dictionaries consist of Key-Value Pairs)
3. Hand this Dictionary over to the Pandas Dataframe

```python
names_list = []
addresses_list = []
expertises_list = []
stars_list = []
ratings_list = []
```

### Loop and Append to List

```python
names_list = [n.text for n in names]
addresses_list = [a.text for a in addresses]
expertises_list = [e.text.replace("\n", ", ") for e in expertises]
stars_list = [s.get_attribute("aria-label") for s in stars]
ratings_list = [r.text for r in ratings]
```

### Data clean up

```python
for i in range(2):
    stars_list.insert(0, "")
    ratings_list.insert(0, "")    
```

```python
print(len(names_list))
print(len(addresses_list))
print(len(expertises_list))
print(len(stars_list))
print(len(ratings_list))
stars_list
```

### Dictionary

```python
dent_dictionary = {"Name":names_list, "Address": addresses_list, "Expertise":expertises_list, "Star":stars_list, "Rating":ratings_list}
```

```python
dent_dictionary
```

### Dictionary to Pandas Dataframe

```python
# Option 1
dentist_df = pd.DataFrame(dent_dictionary)
```

```python
dentist_df
```

### Clean the Data

```python
dentist_df['Star'] = dentist_df['Star'].apply(lambda x: x.replace('star rating', ""))
```

```python
dentist_df.rename(columns={"Star": "Stars"}, inplace=True)
```

```python
dentist_df.replace("", np.nan, inplace=True)
```
